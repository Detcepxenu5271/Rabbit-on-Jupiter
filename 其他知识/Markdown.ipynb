### 1.1 聊一聊计算机语言
从计算理论上说，语言是在某个字符集下可接受的字符串集合。计算机的计算就是建立在对这种语言的操作上。从只有 0/1 的机器语言到强大又复杂的高级语言，各种语言本质上都是不同形式的字符串。

为了进一步理解计算机语言的特点和用途，我们结合自然语言进行一些讨论。假设猫猫早上醒来，发现兔兔向他发送了一条消息“早！”，接下来会发生什么事呢？首先，猫猫的小脑袋从睡迷糊的状态下迅速开机运转，理解了“早！”这个字符串所包含的意思：兔兔醒了，并打了个充满元气的招呼。出于同样的心情，猫猫自然也要回复一句问候，于是他也发送了“早！”。

在上面的例子中有三个过程：首先，猫猫接受了一条字符串形式的信息——也就是语言；然后，猫猫理解了这条信息的内容；最后，猫猫做出了反应，也发送了一条信息。其实，这也正是计算机语言生效的过程：输入、处理和输出。只不过计算机这个人造的产物，其理解语言的过程完全在我们的掌控之中。在 0/1 的底层结构上发生了什么，以及这种底层结构是如何一步步构建上来的，都明明白白地写在计算机科学的发展之路上。而自然语言，虽然我们知道语言长什么样、语言怎么用怎么理解，可是其中的门道似乎让人难以摸清——{\#TODO 这里插入一个语言学的例子}。人类对自然语言的理解，既有从顶层向下的研究，也有构建各种理论、从底层接近自然语言的方法，不过人类对于语言的本质仍然还有很多很多的困惑。

### 1.2 Markdown 是一种好学的标记语言
不同于 Python、C 这样的编程语言，标记语言没有计算的能力，而是用来表示数据的格式。换句话说，就是用一些符号去表示文档的属性、排版等信息。例如，在编写网页的 html 语言中，`<title>兔子洞<\title>`就表示网页的标题叫“兔子洞”，而“兔子洞”三个字并不会显示在网页的页面中，而是被放到了标签栏上，蕴含了“标题”这一指示属性的特殊信息。

标记语言经常被用在网页编写、文档处理（例如用于排版的 LaTeX）中。但是，大型的标记语言往往包含了很多功能，相应也就需要复杂的格式来承载多样的格式变化，这会导致源代码看起来非常吓人，需要显示的原本内容并不突出。例如：
```html
<!-- OSU 主页的一个下载按钮 -->
<div class="user-home__button">
    <a href="https://osu.ppy.sh/home/download" class="btn-home">
        <span class="btn-home__text">下载 osu!</span>
        <span class="btn-home__icon">
            <i class="fas fa-fw fa-download"></i>
        </span>
    </a>
</div>
```

可是，有时候我们会有这样一种需求
1. 需要一些标记格式的方式，让纯文本更加重点突出、层次分明（比如记笔记的时候，想划出重点内容、体现一些层级关系）。
2. 需要在没有相关环境（html 这样的语言是需要浏览器来渲染的）的情况下，面对源代码也能轻松看明白原始内容和格式（让代码更方便使用，不管是专门的编辑器还是 Windows 的记事本都可以做到相对良好的编写体验）。

有没有这样的标记语言呢？有！Markdown 正满足了上面的需求。而这两个需求本身的简单，也为 Markdown 带来了好学的特点。要想完全掌握 Markdown 的语法，一节课就够了！
